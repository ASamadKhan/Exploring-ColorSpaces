```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def plot_3d(pixels, colors_rgb, axis_labels = list("RGB"), axis_limits=((0,255),(0,255),(0,255))):
    """plot pixels in 3d"""
    figure=plt.figure(figsize=(8,8))
    ax = Axes3D(figure)
    
    #set axis limits
    ax.set_xlim(*axis_limits[0])
    ax.set_ylim(*axis_limits[1])
    ax.set_zlim(*axis_limits[2])

    
    #Set axis labels and sizes
    ax.tick_params(axis='both', which='major', labelsize=14, pad=8)
    ax.set_xlabel(axis_labels[0], fontsize=16, labelpad=16)
    ax.set_ylabel(axis_labels[1], fontsize=16, labelpad=16)
    ax.set_zlabel(axis_labels[2], fontsize=16, labelpad=16)

    # Plot pixel values with colors given in colors_rgb
    ax.scatter(
        pixels[:, :, 0].ravel(),
        pixels[:, :, 1].ravel(),
        pixels[:, :, 2].ravel(),
        c=colors_rgb.reshape((-1, 3)), edgecolors='none')

    return ax  # return Axes3D object for further manipulation
    
```

```python
#reading an image
image = cv2.imread("000528.png")
plt.imshow(image)
```

```python
#shape of the image
image.shape
```

```python
# Select a small fraction of pixels to plot by subsampling it
scale = max(image.shape[0], image.shape[1], 64) / 64
small_image = cv2.resize(image,(np.int(image.shape[1] / scale), np.int(image.shape[0] / scale)),interpolation=cv2.INTER_NEAREST)
```

```python
#new Image Shape
small_image.shape
```

```python
# Convert subsampled image to desired color space(s)
img_small_RGB = cv2.cvtColor(small_image, cv2.COLOR_BGR2RGB)  # OpenCV uses BGR, matplotlib likes RGB
img_small_HSV = cv2.cvtColor(small_image, cv2.COLOR_BGR2HSV)
img_small_rgb = img_small_RGB / 255.  # scaled to [0, 1], only for plotting
```

```python
plt.imshow(img_small_rgb)
```

```python
plt.imshow(img_small_HSV)
```

```python
# Plot and show
plot_3d(img_small_RGB, img_small_rgb)
plt.show()
```

```python
plot_3d(img_small_HSV, img_small_rgb, axis_labels=list("HSV"))
plt.show()
```
